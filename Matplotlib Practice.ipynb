```python
xval=np.linspace(-np.pi,np.pi,300)
yval1=np.sin(xval)
yval2=np.cos(xval)
plt.plot(xval,yval1,lw=2,color='red',label='sin(x)')
plt.plot(xval,yval2,lw=2,color='b',label='cos(x)')
plt.title('Trig. Functions')
plt.xlabel('x')
plt.ylabel('sin(x), cos(x)')
plt.axhline(0,lw=0.5,color='k') 
plt.axvline(0,lw=0.5,color='k')
plt.legend()
```

```python
from scipy.stats import gamma
distr=gamma(2.0)
xval=np.linspace(0,10,100)
yval=distr.pdf(xval)
plt.plot(xval,yval,color='orange')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
xmean=distr.mean()
ymean=distr.pdf(xmean)
xmed=distr.median()
ymed=distr.pdf(xmed)
aprops=dict(arrowstyle='->')
plt.annotate('Median={:3.2f}'.format(xmed),(xmed,ymed),(xmed+1,ymed+0.03),arrowprops=aprops)
plt.annotate('Mean={:3.2f}'.format(xmean),(xmean,ymean),(xmean+1,ymean+0.03),arrowprops=aprops)
comment='''In a skewed distribution, 
mean and median usually
have different values.'''
plt.text(4.5,0.15,comment,fontsize=12,color='k')
None
```

```python
from scipy.stats import norm
mmean,msdev=70,4.0
fmean,fsdev=65,3.5
mdist=norm(mmean,scale=msdev)
fdist=norm(fmean,scale=fsdev)
nm,nf=2000,1500
mdata=mdist.rvs(size=nm)
fdata=fdist.rvs(size=nf)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist([mdata,fdata],bins=20,label=['Male','Female'])
plt.legend()
plt.xlabel('Height (in.)')
plt.ylabel('Freq.')
plt.subplot(1,2,2)
plt.boxplot([mdata,fdata],patch_artist=True,labels=['Male','Female'])
plt.ylabel('Height (in.)')
None
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
f=lambda x,y: x**3 -3*x*y**2
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,2,1,projection='3d')
xval=np.linspace(-2,2,100)
yval=np.linspace(-2,2,100)
xgrid,ygrid=np.meshgrid(xval,yval)
zval=f(xgrid,ygrid)
surf=ax.plot_surface(xgrid,ygrid,zval,rstride=5,cstride=5,linewidth=1,cmap='gnuplot')
ax=fig.add_subplot(1,2,2)
plt.contourf(xgrid,ygrid,zval,30,cmap='gnuplot')
fig.colorbar(surf,aspect=18)
plt.tight_layout()
None
```

```python
xval=np.linspace(0,40,500)
def plot_function(ratio=0.1):
    plt.figure(figsize=(10,3))
    yval=np.cos(2*np.pi*xval) + np.cos(2*np.pi*ratio*xval)
    plt.plot(xval,yval,color='g',lw=2)
    plt.axis([0,40,-2.1,2.1])
    plt.show()
    
plot_function()
from ipywidgets import interact 
interact(plot_function,ratio=(0.1,1.1,0.01))
None
```

```python
import Markov_Models as mm
import numpy as np
import matplotlib
matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt
import os
path="/home/ebeyerle/Desktop/PCA/"
os.chdir(path)
mode=1
def plot_fes(mode):
    print('Mode '+str(mode))
    fh=open("mode","w")
    fh.write(str(mode)+"\n")
    fh.close()
    
    with open(path+'anly_'+str(mode)+'.dat','r') as data:
        x=[]
        y=[]
        for line in data:
            p=line.split()
            x.append(float(p[0]))
            y.append(float(p[1]))
    data=[np.column_stack([x,y])]
    model = mm.MSM(data)
    his0 = model.histogram(0, bins=50)
    his1 = model.histogram(1, bins=50)
    his = model.histogram(bins=50)
    ext=model.extent
    bins=50
    x_axis=np.linspace(ext[0],ext[1],bins)
    y_axis=np.linspace(ext[2],ext[3],bins)
    fig=plt.figure(figsize=(10,8))
    plt.contourf(-np.ma.log(his.T/his.sum()), 25, cmap='gnuplot', extent=model.extent)
    plt.xlabel(r'$\theta$ (deg)')
    plt.ylabel(r'$\phi$ (deg)')
    cbar=plt.colorbar()
    cbar.set_label("Free Energy (kcal/mol)")
    plt.show()
interact(plot_fes,mode=(1,150,1))
None
```
